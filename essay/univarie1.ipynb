```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
df = pd.read_csv('../movies.csv')  # Remplacer par le chemin de votre fichier

# Itérer sur les colonnes pour analyser chaque variable
for col in df.columns:
    print(f"Analyse de la variable : {col}")
    print("=================================")
    
    if df[col].dtype in ['int64', 'float64']:  # Variables quantitatives
        print("variable qualitative")
        print(f"Moyenne : {df[col].mean():.2f}")
        print(f"Médiane : {df[col].median():.2f}")
        print(f"Écart type : {df[col].std():.2f}")
        print(f"Valeur min : {df[col].min()}, Valeur max : {df[col].max()}")
        print(f"Asymétrie : {df[col].skew():.2f}, Aplatissement : {df[col].kurt():.2f}")
        
        # Création de graphiques
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))
        sns.histplot(df[col], kde=True, ax=axes[0])
        axes[0].set_title(f"Histogramme de {col}")
        
        sns.boxplot(x=df[col], ax=axes[1])
        axes[1].set_title(f"Boxplot de {col}")
        
        plt.show()
    
    elif df[col].dtype == 'object':  # Variables qualitatives
        freq_table = df[col].value_counts()
        print("variable qualitative")
        print("Fréquences :")
        print(freq_table)
        
        # Création d'un graphique
        plt.figure(figsize=(8, 4))
        sns.barplot(x=freq_table.index, y=freq_table.values)
        plt.title(f"Diagramme en barres de {col}")
        plt.xlabel(col)
        plt.ylabel("Fréquence")
        plt.xticks(rotation=45)
        plt.show()
    
    print("\n")
```
