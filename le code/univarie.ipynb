```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

le but de l'analyse univarie est de analysez chaque variable individuelmment

```python
data = pd.read_csv('movies.csv')
data = data.drop(columns=['Gross']) 
print (data.info())
data = data.dropna(axis=0)
print (data.info())
data = data.drop_duplicates()
print (data.info())
data.to_excel('CleanData.xlsx')
```

```python
for col in data.columns:
    print(f"Analyse de la variable : {col}")
    print("=================================")
    
    if data[col].dtype in ['int64', 'float64']:  # Variables quantitatives
        print("variable qualitative")
        print(f"Moyenne : {data[col].mean():.2f}")
        print(f"Médiane : {data[col].median():.2f}")
        print(f"Écart type : {data[col].std():.2f}")
        print(f"Valeur min : {data[col].min()}, Valeur max : {data[col].max()}")
        print(f"Asymétrie : {data[col].skew():.2f}, Aplatissement : {data[col].kurt():.2f}")
        
        # Création de graphiques
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))
        sns.histplot(data[col], kde=True, ax=axes[0])
        axes[0].set_title(f"Histogramme de {col}")
        
        sns.boxplot(x=data[col], ax=axes[1])
        axes[1].set_title(f"Boxplot de {col}")
        
        plt.show()
    elif data[col].dtype == 'object':  # Variables qualitatives
        freq_table = data[col].value_counts()
        print("variable qualitative")
        print("Fréquences :")
        print(freq_table)
        
        # Création d'un graphique
        plt.figure(figsize=(16, 8))
        sns.barplot(x=freq_table.index, y=freq_table.values)
        plt.title(f"Diagramme en barres de {col}")
        plt.xlabel(col)
        plt.ylabel("Fréquence")
        plt.xticks(rotation=45)
        plt.show()
    
    print("\n")
   
    
```
